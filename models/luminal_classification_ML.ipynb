```python
import numpy as np
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# X = features, y = target
X = df.drop(columns=["Luminal", "image_name"])      # replace "y" with your target column name
y = df["Luminal"]

# Lasso with internal CV + standardization
lasso = make_pipeline(
    StandardScaler(),
    LassoCV(cv=5, random_state=0, max_iter=10000, tol=1e-4, alphas=np.logspace(-4, 4, 50))
)

lasso.fit(X, y)

# get trained LassoCV estimator from the pipeline
lasso_cv = lasso.named_steps["lassocv"]

# boolean mask of selected (nonâ€‘zero) coefficients
coef = lasso_cv.coef_
selected_mask = coef != 0

# corresponding feature names
selected_features = X.columns[selected_mask].tolist()
print("Selected features:", selected_features)

# dataframe with only selected features
df_selected = df[selected_features + ["Luminal"]]   # keep target too
```

```python
df_selected["image_name"] = df["image_name"]
df_selected
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.inspection import permutation_importance

# Features/target
X = df_selected.drop(columns=["Luminal", "image_name"])
y = df_selected["Luminal"]

# ----- balanced 20 positive / 20 negative for training -----
# indices for each class
pos_idx = y[y == 1].index
neg_idx = y[y == 0].index

# randomly sample 20 from each (assumes you have â‰¥20 of each)
pos_train_idx = np.random.choice(pos_idx, size=25, replace=False)
neg_train_idx = np.random.choice(neg_idx, size=25, replace=False)

train_idx = np.concatenate([pos_train_idx, neg_train_idx])

X_train = X.loc[train_idx]
y_train = y.loc[train_idx]

# use the remaining cases as test set
test_idx = y.index.difference(train_idx)
X_test = X.loc[test_idx]
y_test = y.loc[test_idx]
# -----------------------------------------------------------

# Same SVM model as before
svm_clf = make_pipeline(
    StandardScaler(),
    SVC(kernel="rbf", C=1.0, gamma="scale", probability=True, random_state=0)
)

svm_clf.fit(X_train, y_train)

# Permutation importance on heldâ€‘out test set
result = permutation_importance(
    svm_clf, X_test, y_test,
    n_repeats=30,
    random_state=0,
    n_jobs=-1
)

importances_mean = result.importances_mean
importances_std = result.importances_std
feature_names = X.columns

# Sort features by importance (descending)
indices = np.argsort(importances_mean)[::-1]

print("Feature importances (permutation, decrease in score):")
for idx in indices:
    print(
        f"{feature_names[idx]}: "
        f"mean={importances_mean[idx]:.4f}, std={importances_std[idx]:.4f}"
    )
```

```python
y_pred = svm_clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# AUC (use probability of positive class)
y_proba = svm_clf.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_proba)
print("ROC AUC:", auc)

fpr, tpr, thresholds = roc_curve(y_test, y_proba)
```

```python
import numpy as np

# df: your original dataframe
# target column name (if present)
target_col = "Luminal"     # change as needed

# use only feature columns
feature_cols = [c for c in df.columns if c != target_col]
X = df[feature_cols]
X = df.drop(columns=["image_name"])

# compute Pearson correlation matrix between features
corr_matrix = X.corr(method="pearson").abs()

# keep only upper triangle of matrix (to avoid duplicate pairs)
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

# set correlation threshold for â€œtoo similarâ€ features
threshold = 0.6

# columns to drop: any feature that has correlation > threshold with a previous feature
to_drop = [
    column for column in upper.columns
    if any(upper[column] > threshold)
]

print("Dropping correlated features:", to_drop)

# dataframe with only useful (nonâ€‘redundant) columns
useful_features = [c for c in feature_cols if c not in to_drop]
df_corr_selected = df[useful_features + ([target_col] if target_col in df.columns else [])]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.inspection import permutation_importance

# Features/target
X = df_corr_selected.drop(columns=["Luminal", "image_name"])
y = df_corr_selected["Luminal"]

# ----- balanced 20 positive / 20 negative for training -----
# indices for each class
pos_idx = y[y == 1].index
neg_idx = y[y == 0].index

# randomly sample 20 from each (assumes you have â‰¥20 of each)
pos_train_idx = np.random.choice(pos_idx, size=30, replace=False)
neg_train_idx = np.random.choice(neg_idx, size=30, replace=False)

train_idx = np.concatenate([pos_train_idx, neg_train_idx])

X_train = X.loc[train_idx]
y_train = y.loc[train_idx]

# use the remaining cases as test set
test_pos_idx_set = y[y == 1].index.difference(train_idx)
test_neg_idx_set = y[y == 0].index.difference(train_idx)

pos_test_idx = np.random.choice(test_pos_idx_set, size=20, replace=False)
neg_test_idx = np.random.choice(test_neg_idx_set, size=10, replace=False)

test_idx = np.concatenate([pos_test_idx, neg_test_idx])

X_test = X.loc[test_idx]
y_test = y.loc[test_idx]
# -----------------------------------------------------------

# Same SVM model as before
svm_clf = make_pipeline(
    StandardScaler(),
    SVC(kernel="rbf", C=2.0, gamma="scale", probability=True, random_state=56)
)

svm_clf.fit(X_train, y_train)

# Permutation importance on heldâ€‘out test set
result = permutation_importance(
    svm_clf, X_test, y_test,
    n_repeats=30,
    random_state=233,
    n_jobs=-1
)

importances_mean = result.importances_mean
importances_std = result.importances_std
feature_names = X.columns

# Sort features by importance (descending)
indices = np.argsort(importances_mean)[::-1]

print("Feature importances (permutation, decrease in score):")
for idx in indices:
    print(
        f"{feature_names[idx]}: "
        f"mean={importances_mean[idx]:.4f}, std={importances_std[idx]:.4f}"
    )
```

```python
y_pred = svm_clf.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# AUC (use probability of positive class)
y_proba = svm_clf.predict_proba(X_test)[:, 1]
auc = roc_auc_score(y_test, y_proba)
print("ROC AUC:", auc)

fpr, tpr, thresholds = roc_curve(y_test, y_proba)
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC

n_repeats = 50
auc_scores = []

for seed in range(n_repeats):
    # repeat your balanced sampling each time
    np.random.seed(seed)

    # X, y as before
    X = df_corr_selected.drop(columns=["Luminal", "image_name"])
    y = df_corr_selected["Luminal"]

    pos_idx = y[y == 1].index
    neg_idx = y[y == 0].index

    pos_train_idx = np.random.choice(pos_idx, size=30, replace=False)
    neg_train_idx = np.random.choice(neg_idx, size=30, replace=False)
    train_idx = np.concatenate([pos_train_idx, neg_train_idx])

    X_train = X.loc[train_idx]
    y_train = y.loc[train_idx]

    test_pos_idx_set = y[y == 1].index.difference(train_idx)
    test_neg_idx_set = y[y == 0].index.difference(train_idx)

    pos_test_idx = np.random.choice(test_pos_idx_set, size=20, replace=False)
    neg_test_idx = np.random.choice(test_neg_idx_set, size=10, replace=False)
    test_idx = np.concatenate([pos_test_idx, neg_test_idx])

    X_test = X.loc[test_idx]
    y_test = y.loc[test_idx]

    svm_clf = make_pipeline(
        StandardScaler(),
        SVC(kernel="rbf", C=2.0, gamma="scale", probability=True, random_state=56)
    )
    svm_clf.fit(X_train, y_train)

    y_proba = svm_clf.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_proba)
    auc_scores.append(auc)

auc_scores = np.array(auc_scores)
print("AUC mean:", auc_scores.mean())
print("AUC std :", auc_scores.std())
print("AUC min :", auc_scores.min())
print("AUC max :", auc_scores.max())
```
